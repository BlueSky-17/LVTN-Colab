```python
import regex as re

txt = """

  Biểu mẫu xác nhận thông tin cư trú bao gồm: Theo đề nghị của Ông/Bà, Số định danh cá nhân,
  Họ chữ đệm và tên của Ông/Bà, Ngày, tháng, năm sinh, Giới tính, Số định danh cá nhân, Dân tộc,
  Tôn giáo, Quê quán, Nơi đăng ký khai sinh, Nơi thường trú, Nơi tạm trú, Nơi ở hiện tại, Họ, chữ
  đệm và tên chủ hộ, Quan hệ với chủ hộ, Số định danh cá nhân chủ hộ, Thông tin các thành viên
  khác trong hộ gia đình, Nội dung xác nhận khác.


"""

txt1 = """

Đăng ký tham dự ngày hội việc làm CSE Job Fair năm học 2024:
  + Họ và tên
  + Mã số sinh viên
  + Khoa/Văn phòng/Trung tâm (Máy tính, Điện, Hóa, Xây dựng)
  + Email
  - Số điện thoại
  - Câu hỏi gửi đến BTC.
  + Giới tính (Nam, Nữ)

"""

# res = re.search(r'(?i)[\p{L}\d/,.:+\- ]+', txt1, re.DOTALL)
res = re.search(r'(?=.*\+)(?=.*\-)(?=.*:)', txt1, re.DOTALL)

if res:
  print("Matching")
  print(res)
else:
  print("Not Matching")
  print(res)



tieude = re.findall(r'(?i)[\p{L}\d/,. ]+(?=:)', txt1, re.DOTALL)
print(f'Tiêu đề: "{tieude[0]}"')

tokenize = re.findall(r'(?i)(?<=[+-])[\p{L}\d/,. ]+', txt1, re.DOTALL)
# tokenize.pop(0)
# print(tokenize)

print("Các câu hỏi: [")

for ques in tokenize:
  print(f'"{ques}",')

print("]")
```

```python
from transformers import AutoTokenizer, AutoModelForTokenClassification
from transformers import pipeline


tokenizer = AutoTokenizer.from_pretrained("NlpHUST/ner-vietnamese-electra-base")
model = AutoModelForTokenClassification.from_pretrained("NlpHUST/ner-vietnamese-electra-base")

nlp = pipeline("ner", model=model, tokenizer=tokenizer)
example = "Liên quan vụ việc CSGT bị tố đánh dân, trúng một cháu nhỏ đang ngủ, đang lan truyền trên mạng xã hội, Đại tá Nguyễn Văn Tảo, Phó Giám đốc Công an tỉnh Tiền Giang vừa có cuộc họp cùng Chỉ huy Công an huyện Châu Thành và một số đơn vị nghiệp vụ cấp tỉnh để chỉ đạo làm rõ thông tin."

ner_results = nlp(example)

for obj in ner_results:
  print(obj)
```
